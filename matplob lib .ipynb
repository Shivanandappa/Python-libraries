```python
# compute the x and y co ordinates for points on a sine curve

x = np.arange(0,4 * np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")

#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
# subplot
# compute the x and y cordinates for points on sine and cosine curves
x = np.arange(0,5 * np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 

plt.subplot(2,1,1)

#Make the first plot 
plt.plot(x,y_sin,'r--')
plt.title('Sine')

# set the Second subplot as active, and make the second plot. 
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')

# show the figure 
plt.show()
```

```python
## Barplot

x = [2,8,10]
y = [11,16,9]

x2= [3,9,11]
y2 = [6,15,7]

plt.bar(x,y,align = 'center')
plt.bar(x2, y2, color = 'g')
plt.title('Bar graph')
plt.ylabel('y axis')
plt.xlabel('x axis')

plt.show()
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]

#rectangular box plot
plt.boxplot(data,vert = True,patch_artist = True);
```

```python
# Data to plot
labels = 'python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.4,0,0.2,0) # explode 1st slice 

# Plot ( to change the colors you need to see the list of colour assigned)
plt.pie(sizes, explode= explode, labels = labels, colors = colors,
autopct = '%1.1f%%', shadow = True)

plt.axis('equal')
plt.show()
```
